```python
import pandas as pd

df = pd.read_csv('CustomerSurvey.csv')
```

```python
# Calculate the percentage of null values column-wise
null_percentage = (df.isnull().sum(axis=0) / df.shape[0]) * 100
```

```python
null_percentage_sorted = null_percentage.sort_values(ascending=False)
```

```python
# Count the number of columns with more than 20% null values
columns_with_more_than_20_percent_nulls = null_percentage[null_percentage > 7].count()

print("Number of columns left:", 105-columns_with_more_than_20_percent_nulls)
```

```python
# Filter columns with less than or equal to 5% null values
selected_columns = null_percentage[null_percentage <= 7].index

# Remove columns with more than 5% null values
df_filtered = df[selected_columns]
```

```python
# Calculate the percentage of null values row-wise
null_percentage1 = (df_filtered.isnull().sum(axis=1) / df_filtered.shape[1]) * 100
```

```python
# Filter rows with less than or equal to 20% null values
final_filtered_df = df_filtered[null_percentage1 < 4]
```

```python
# Calculate the total number of null values in the DataFrame
total_null_values = final_filtered_df.isnull().sum()
```

```python
cleaned_df = final_filtered_df
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define the occupation data
data = {
    'Accountants': 7,
    'Lawyer': 3,
    'General Purpose contractor': 1,
    'Design': 10,
    'Educator': 43,
    'Homemaker': 17,
    'Business Professional': 114,
    'Student': 10,
    'Retired': 4
}

# Generate word cloud
wordcloud = WordCloud(width=400, height=100, background_color='white').generate_from_frequencies(data)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
